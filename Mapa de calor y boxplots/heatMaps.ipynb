```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
dia = pd.read_csv('Diamonds are forever.csv', na_values = '')
dia
```

```python
sns.set(rc={'figure.figsize':(11.7, 8.27)})
sns.distplot(dia['Carat'], bins=30)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(11.7, 8.27)})
sns.distplot(dia['Colour'], bins=6)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(11.7, 8.27)})
sns.distplot(dia['Clarity'], bins=5)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(11.7, 8.27)})
sns.distplot(dia['Lab'], bins=3)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(11.7, 8.27)})
sns.distplot(dia['Price ($US)'], bins=50)
plt.show()
```

```python
correlation_matrix = dia.corr().round(2)

sns.heatmap(data = correlation_matrix, annot=True)
```

```python
plt.boxplot(dia['Price ($US)'], notch=True)
plt.show()
```
